```python
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
sensor_df = train_df.drop(
    columns=[
        'engine_id',
        'time',
        'operational_setting_1',
        'operational_setting_2',
        'operational_setting_3',
        'time_until_fault',
        'time_until_failure'
    ]
)
```

```python
sensor_df.head() 
```

```python
sensor_df_norm = (sensor_df-sensor_df.mean())/sensor_df.std()
```

```python
sensor_df_norm
```

```python
sensor_df_norm = sensor_df_norm.drop(
    columns=[
        'T2',
        'Nf_dmd',
        'PCNfR_dmd',
    ]
)
```

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca = PCA(n_components=1)
health_indicators = pca.fit_transform(sensor_df_norm)
health_indicators = (health_indicators - np.min(health_indicators)) / (np.max(health_indicators) - np.min(health_indicators))
```

```python
health_indicators.shape
```

```python
train_df['health_indicator'] = health_indicators
```

```python
import matplotlib.pyplot as plt

engine = 1
plt.plot(train_df[train_df.engine_id==engine].health_indicator.to_numpy(), label='normalize health indicator')
plt.title('Health Indicator over Engine Life')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = train_df.health_indicator.to_numpy().reshape(-1,1)
y = train_df.time_until_failure
model = LinearRegression().fit(X, y)
model.score(X,y)
```

```python
pred = model.predict(X)
```

```python
train_df['estimated_rul'] = pred
```

```python
engine = 1
plt.plot(train_df[train_df.engine_id==engine].estimated_rul.to_numpy(), label='estimated rul')
plt.plot(train_df[train_df.engine_id==engine].time_until_failure.to_numpy(), label='ground truth')
plt.legend()
plt.title("Estimated RUL vs Ground Truth")
plt.show()
```

```python
train_df.to_csv('train.csv')
```
